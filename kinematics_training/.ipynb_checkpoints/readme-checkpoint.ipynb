# 運動・動作の分析
## 動作の原点
### アフォーダンス
　ギブソンは感覚器官から知覚を説明する知覚理論のパラドクスを指摘し，「基礎的定位づけシステム」「視るシステム」「聴くシステム」「味わい・嗅ぐシステム」「接触のシステム」を持つ「知覚システム」として表現し，動物の行為が身体内部のみで決定されるのではなく，環境が行為や運動制御に大きく関与していることを提示した。<br>
　知覚者はを刺激に対する単なる受動的存在ではなく，自ら積極的に環境の中の情報を能動的に探索できる主体であると考える。<br>
　環境が動物に提供する環境の性質・情報による行為の可能性をアフォーダンスという言葉を用いて表現し，加えて，動物が取り囲む環境に適応し行為を行うためには，意識・無意識かわ問わず，アクティブタッチの重要性を示唆する。
#### 知覚循環
　課題ー環境ー身体は制約しつつ行為は成り立つ。
 > 1.タスクのゴール=課題<br>
 > 2.タスク・ダイナミクスを特定，あるいは制約する規則（個体発生学的な制限）=身体<br>
 > 3.タスク・ダイナミクスを特定，あるいは制約する道具だてや装置=環境<br>
### クラインフォーゲルバッハの運動学
　脊椎動物である人間は内骨格構造とそれをつなぐ外側の筋により動く。支持面と接する骨格は球形や円形で傾きやすく，関節面の多くは弯曲し摩擦はほとんどない。関節は筋や靭帯などの弾力性のある構造で支えられ，柔軟性があり大きな自由度を持つため骨格だけでは安定性を確保できない。つまり，かなり不安定な構造をしている。横紋筋の発生は強く速い運動を可能としたが，姿勢制御にも働く必要があり神経系による協応が不可欠である。<br>
　クラインフォーゲルバッハは身体体節を５つに分類している。両側股関節の中心と肩関節を結んだ対角線の交点付近を中枢とよび，そこから遠く離れた頭部や四肢を末梢と定義している。<br>
　人間は重くて丸い頭部や長くて重い上下肢が身体の末梢に位置するため，これらの身体の体節の抗重力活動における安定性が，中枢を形成する体幹部に求められることになる。<br>

## 動作・姿勢の表し方
### 姿勢について
#### 構えと体位
##### 構え
頭部・体幹・四肢の身体各部位の相対的な位置関係を意味するもの。<br>頭部前屈位，上肢外転位，体幹後屈位といった形で表現される。 <br> 構えは体節相互の位置関係を関節角度測定によって表示することが可能である。このときには立位であるか，背臥位であるかは問題にしない。
##### 体位
身体が重力の方向とどのような関係にあるかを示すもの。<br>
立位，背臥位，側臥位などと表される。<br>
人間の立位姿勢は重力が上下方向，頭尾方向に働いている状態であり，背臥位では重力は前後方向に働いている。<br>

#### 静的姿勢と動的姿勢
機能的立場から，姿勢は静的姿勢と動的姿勢の二つに分ける。運動は構えを意味する姿勢が時間的・空間的に連続して変化したものである。運動は身体の重力方向との関係を示す体位と，身体各部の相対的位置関係を示す構えの変化の二つの要因を含む。動的姿勢はある動作の起始となる準備状態，あるいは一連の動作の時間軸上の１点を観察したものである。一方，動作と姿勢との関連について，時間軸における制限を加味して，動作のある瞬間を捉えたものは姿勢とは言い得ず，ある時間は維持される身体の重力方向との関係と，身体各部の相互の位置的関連性を示すものが姿勢であるとする考え方もある。静的姿勢，動的姿勢に関する解釈は多様で一定でない。立位姿勢を保持するとき，外観からは静止状態を保っているように見えるが，わずかな重心位置の移動が常にある。身体が厳密に単一の姿勢を長時間維持することはない。
### 動作について
人間の行動は運動，動作，行為にわけることができる。
>例えば・・・<br>
>眼球の動きについて，水平方向に5度動く：運動<br>
>視線の移動：動作<br>
>「Aが話の終わりにちつもBを見る，これはAが話を終わり，Bが話し始める合図である」：行為<br>

理学療法では運動療法を大きな武器としている。適切な運動を行わせる，指導することで対象者の運動様式に変化を与えることで対象者と環境の関係作りに変化を与えることを仕事とする。ということは動作を観察し，分析する能力が必然的に求められる。
#### 方法
基本原則は次の通りであるが，
>1. 分析する運動を正しく記載する。
>2. 運動の各相は関節運動と筋活動の分析によって決定する。
>3. 運動は分析のための一定の評価基準で決定する。

##### 例
>1. 動作名<br>
>2. 相決め<br>
>3. 記録：開始姿勢〜動作の記録〜終了姿勢

## 重心動揺計
### 計測の原理

> ある３点の圧力センサの上に，一枚のプレートを置く。その上にある物体の圧中心点をあらかじめ決めた座標上で求める。 <br>
> それを計測し，図で表す，揺れた大きさを計算する，揺れた速度を計算するなど，様々な解析を行う。

### 計測方法
> デモンストレーションする。

### 計測内容
> 全員の（チャレンジ！！）<br>
> 開眼閉脚立位<br>
> 閉眼閉脚立位<br>をペア（被検者とPC操作者）になって計測する。<br>

### 計測データの解析
> 以下にプログラムを例示する。<br><br>
> 1. データを読み込む。<br>
> 2. 図示してデータの中身を確認する（ここでうまく表示されなければやり直し）。<br>
> 3. フィルタリングをする。<br>
> 4. 共通する座標系に置き換える（方向を決定する）。ここでは，主軸変換をする。<br>
> 5. 一度図示して回転方向を決める。<br>
> 6. 回転角度を算出する。<br>
> 7. 矩形面積を求める。<br>
> 8. 総軌跡長を求める。<br>
